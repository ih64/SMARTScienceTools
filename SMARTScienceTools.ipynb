# SMARTScienceTools Tutorial

This tutorial will quickly show some of the abilities of the SMARTScienceTools package.
In order to work on your system, it requires astropy, matplotlib, and urllib2. If you run this ipython notebook from github, you should have no problems

## Getting Started

sst consists of a class, blazar, and a python list which holds the names of all available blazars. 
It is imported like any python module.

```python
%matplotlib inline
import sst
```

To crate an instance of the class, we just need to pass in the name of a blazar as a string

```python
target=sst.blazar('1510-089')
```

To get all the photometric data of this target, we can use the getdata method. getdata() sends a request to the smarts website for the csv file associated with this target. It is then parsed with astropy ascii table and returned. The nan entries represent cells where data was not taken at a particular date or at a particular band.

```python
data=target.getdata()
print data[100:105]
```

We can work on the data for a particualr column the way we usually would with any astropy ascii table. You can find more information on astropy ascii tables here: http://astropy.readthedocs.org/en/latest/io/ascii/

```python
print data['bdate']
```

## Other Methods 

If you are not sure if your target is available with sst, you can quickly check by using the doiexist method

```python
system=sst.blazar('11235')
system.getdata()
```

```python
system.doiexist()
```

The rollcall method will print the names of all the targets that can be used

```python
system.rollcall()
```

## Making plots

We can create some quick plots using sst. To make light curves, use the lightcurve method

```python
target.lightcurve(data)
```

Light curves of particular bands can be created as well. lightcurve() has a keyword, filters, which is a python list ['b', 'v', 'r', 'j', 'k'] by default (all of the filters used to monitor blazars with  SMARTS). You can pass a list consisting of any combination of these filters as well, to plot particular filters.

```python
target.lightcurve(data,filters=['b','j'])
```

It is also possible to make color magnitude diagrams

```python
target.cmd(data)
```

The user can define the bluer filter and redder filter using the keywords. By default, blue='b' and red='r'. We can specify blue='r' and red='k' to crate a different color magnitude diagran, for example.

```python
target.cmd(data,blue='r',red='k')
```

```python
target.cmd(data,blue='p',red='k')
```

```python
target.cmd(data,blue='r',red='p')
```

```python
target.cmd(data,blue='p',red='p')
```

```python
target.lightcurve(data, filters=5)
```
